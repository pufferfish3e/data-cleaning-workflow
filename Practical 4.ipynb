# Annex: Creating dataframes

```python
import pandas as pd
import numpy as np

# Define job titles and corresponding salary ranges
job_titles = ['Data Scientist', 'Software Engineer', 'Data Analyst', 'Project Manager', 'Marketing Specialist', 'Web Developer']
salary_ranges = [(60000, 120000), (70000, 140000), (50000, 100000), (65000, 130000), (45000, 90000), (55000, 110000)]

# Create a list to store the data
data = []

# Generate data for 100 job listings
for i in range(100):
    job_title = np.random.choice(job_titles)  # Randomly select a job title
    salary_range = salary_ranges[job_titles.index(job_title)]  # Get salary range for the job title

    # Introduce missing value with 15% probability
    if np.random.rand() < 0.15:
        salary = None  # Set salary to NaN (missing)
    else:
        salary = np.random.randint(salary_range[0], salary_range[1])  # Generate random salary

    data.append([job_title, salary])

# Create a pandas DataFrame
df = pd.DataFrame(data, columns=['Job Title', 'Salary'])

# Display the DataFrame
print(df.head())
```
