# Question: Music Data Analytics

You are a music analyst tasked with building a music recommendation system. You are given data in the form of a list of tuples. Each tuple represents a user and their music preferences, structured as follows:

(user_id, [(artist, genre, rating), (artist, genre, rating), ...])

```python
#run this cell
music_data = [
    (1, [("ArtistA", "Rock", 4), ("ArtistB", "Pop", 5), ("ArtistC", "Rock", 3)]),
    (2, [("ArtistB", "Pop", 2), ("ArtistD", "Jazz", 4), ("ArtistE", "Classical", 5)]),
    (3, [("ArtistA", "Rock", 5), ("ArtistD", "Jazz", 3), ("ArtistF", "Pop", 4)])
]
```

# Your tasks is to complete the three different functions based on the requirements:


### 1. get_genre_preferences(user_id, music_data):
*   This function takes a user_id and the music_data as
input.
* It should return a dictionary where keys are genres and values are the average rating given by the user to that genre.
* If the user has not rated any songs in a particular genre, the genre should not be included in the dictionary.
* If the user_id is not found in the data, raise a ValueError with the message "User not found".

**Example usage: genre_preferences = get_genre_preferences(1, music_data)  -> Output: {'Rock': 3.5, 'Pop': 5}**

This is because the index 1 returns this data (1, [("ArtistA", "Rock", 4), ("ArtistB", "Pop", 5), ("ArtistC", "Rock", 3)]).
The average score of Rock is 4+3/2 and Pop is 5/1

```python
def get_genre_preferences(user_id, music_data):
    genre_preferences = {}
    hashmap = {}
#Write your code here
    # Loop through the elements of music_data with a for loop.
    for i in range(len(music_data)):
        # Find the music_data element associated with the user_id.
        if user_id == music_data[i][0]:
            # Loop through the ratings of artists within the row.
            for j in range(len(music_data[i][1])):
                # Use a hashmap to store occurences of genres.
                # Check if genres exist in the hashmap.
                if music_data[i][1][j][1] not in hashmap:
                    # If they do not exist, create a new instance of a genre with an array as a value storing the rating and the number of occurences.
                    hashmap[f"{music_data[i][1][j][1]}"] = [music_data[i][1][j][2],1]
                else:
                    # If they exist, update the instance of the genre with the total rating and the total number of occurences of the genre.
                    hashmap[f"{music_data[i][1][j][1]}"][0] += music_data[i][1][j][2]
                    hashmap[f"{music_data[i][1][j][1]}"][1] += 1

            # Lopp through the hashmap and construct the result by assigning the genre as the key and the average rating by taking value[1] / value[2] as the value.
            for k in hashmap.keys():
                genre_preferences[k] = hashmap[k][0] / hashmap[k][1]
            return genre_preferences
    # Raise a ValueError if the user is not found.
    else:
            raise ValueError("User not found.")
#run function
get_genre_preferences(1, music_data)
```

### 2. recommend_artists(user_id, music_data):

This function takes a user_id and the music_data as input. It should recommend artists to the user based on their preferred genres. The function should:

- Get the user's genre preferences using get_genre_preferences.
- Find other users who have rated songs in the same genres with a rating of 4 or higher.
Recommend artists that these other users have rated highly (4 or higher) but that the given user has not yet rated.
- Return a list of recommended artists.
- If no recommendations can be made (e.g., the user has rated all artists or there are no other users with similar preferences), return an empty list.
- If the user_id is not found, raise a ValueError with the message "User not found".

**Example: recommendations = recommend_artists(1, music_data)  -> Output: ['ArtistF']**

This is because index 1's liking is Pop as mentioned from the previous function, and ArtistF is someone that index 1 has not listen to, and therefore a recommendation.

```python

def recommend_artists(user_id, music_data):
    results = []
#Write your code here
    # Get the user's genre preferences by calling the get_genre_preferences function to obtain the genre preferneces of the artist.
    genre_preferences = get_genre_preferences(user_id=user_id, music_data=music_data)
    # Filter the input dictionary using dictionary comprehension to drop elements with a value greater than 4.
    genre_preferences = {k:v for k,v in genre_preferences.items() if v >= 4}
    # Loop through the music data and filter out the current user.
    for i in range(len(music_data)):
        # Skip if user has been used for getting genre preferences
        if user_id-1 != i:
            # Inner for loop to loop through Artists
            for j in music_data[i][1]:
                # Add to list if conditions are satisfied
                if j[1] in genre_preferences and j[2] >= 4:
                    results.append(j[0])
    return results
    ###Write more code to solve the problem

    
#run function
recommend_artists(1, music_data)
```
