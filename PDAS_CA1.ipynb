#### Handle Missing Values Strategy

Our missing value strategy will help us to minimize data loss, while help us get closer to data that can answer our research question.

**Salary Columns (max_salary, med_salary, min_salary) - DROP rows with missing values.**

-   These are our **primary variables** for the research question
-   We will not take the median, as that would impute fake data which might ruin our dataset
-   After dropping, we retain records with at least one salary value for meaningful analysis

After further analysis, we decided on a more revised approach. We found that dropping all rows with a NaN value would essentially be dropping the entire dataset.

-   Using `how="all"` keeps rows with **at least one** salary value.
-   Most jobs report either max_salary OR med_salary, not both
-   This prevents losing all data
-   Rows with at least one salary can still show variation by work type

**Categorical Features (pay_period, formatted_work_type, compensation_type, currency) - DROP rows with missing values.**

-   **pay_period** and **formatted_work_type** are essential grouping variables - without them we cannot categorize jobs by work type/status
-   **currency** is critical for standardization - mixed or unknown currencies make salary comparisons invalid
-   **compensation_type** helps validate that we're comparing equivalent compensation structures

**Remote_allowed - RETAIN rows with missing values:**

-   This is a supplementary feature, not central to the research question
-   Missing values likely indicate "unspecified" rather than true missing data
-   Can be treated as a separate category or analyzed separately without compromising core analysis

**Final Strategy:**

1. Drop rows missing any salary value (max, med, or min)
2. Drop rows missing pay_period, formatted_work_type, compensation_type, or currency
3. Retain rows with missing remote_allowed for maximum data retention
4. This ensures all remaining records have complete information for our primary analysis

# Step 4: Data Preprocessing

Now that we have cleaned our data and handled missing values, we need to preprocess the dataset to prepare it for analysis and visualization. This step involves transforming and standardizing variables to ensure consistency and comparability.

## Currently, our dataset has:

-   **36,073 clean records** with no missing values

## Preprocessing Objectives:

### 1. **Salary Standardization**

-   Currently we have salaries in both HOURLY and YEARLY pay periods
-   We need to standardize all salaries to a common time period (YEARLY) for fair comparison
-   Formula: `yearly_salary = hourly_salary × 2080` (assuming 40 hours/week × 52 weeks)

### 2. **Currency Verification**

-   Verify all salaries are in the same currency (USD)
-   Check for any non-USD entries that need conversion or removal

### 3. **Categorical Data Standardization**

-   Examine unique values in **formatted_work_type** (Full-time, Part-time, Contract, etc.)
-   Examine unique values in **remote_allowed** (including our "Unspecified" category)
-   Examine unique values in **compensation_type** to ensure consistency
-   Standardize any inconsistent formatting (capitalization, spacing, etc.)

### 4. **Outlier Detection**

-   Identify and handle extreme salary values that may skew our analysis
-   Use statistical methods (IQR, z-score) to detect outliers

These preprocessing steps will certify that our data is ready for visualization and analysis to answer our research question: **"How do salary offerings vary by employment type and remote work arrangement?"**


We will first check for unique values within our dataset for the salary column just to make sure all our data is ready for analysis.


````python
# Check for the type of conversions we can convert our money into.
# We will standardize by converting all to year.
df["pay_period"].unique()
````

````python
# Assumptions: 
# 9-5 work week (8 hours/day, Monday-Friday) = About 40 hours/week.
# Formulas:
# Hours Worked per Year: 8 * 5 * 52 = 2080
# Hence, Yearly = Hourly * 2080
# Weekly to Yearly = Weekly * 52
# Biweekly to Yearly = Biweekly * 26
# Monthly to Yearly = Monthly * 12
df.loc[df["pay_period"] == "HOURLY","med_salary"] = df.loc[df["pay_period"] == "HOURLY","med_salary"] * 2080
df.loc[df["pay_period"] == "WEEKLY","med_salary"] = df.loc[df["pay_period"] == "WEEKLY","med_salary"] * 52
df.loc[df["pay_period"] == "BIWEEKLY","med_salary"] = df.loc[df["pay_period"] == "BIWEEKLY","med_salary"] * 26
df.loc[df["pay_period"] == "MONTHLY","med_salary"] = df.loc[df["pay_period"] == "MONTHLY","med_salary"] * 12
# Drop feature since we don't need it
df = df.drop("pay_period", axis=1)
````

Now, we will check for outliers using the interquartile range (only applies to salary).


````python
# Calculate IQR for med_salary to detect outliers
Q1 = df['med_salary'].quantile(0.25)
Q3 = df['med_salary'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Q1 (25th percentile): ${Q1:,.2f}")
print(f"Q3 (75th percentile): ${Q3:,.2f}")
print(f"IQR: ${IQR:,.2f}")
print(f"Lower bound: ${lower_bound:,.2f}")
print(f"Upper bound: ${upper_bound:,.2f}")

# Identify outliers
outliers = df[(df['med_salary'] < lower_bound) | (df['med_salary'] > upper_bound)]
print(f"\nNumber of outliers detected: {len(outliers)}")
print(f"Percentage of outliers: {(len(outliers)/len(df)*100):.2f}%")
````

## Visualization 3: Heatmap of Salary by Employment Type and Work Status

This heatmap visualizes the amount of salary by employment type by adding color to salary, where cooler colors mean lower salary, and warmer colors mean higher salary. We can also compare to work status and see a side by side comparison on how they differ. This also provides us with the numerical value (that the previous graph could not provide) so we can get a better understanding of the median salaries by work type.


````python
# Create a color palette table defined with custom colors.
catppuccin_colors = [
    PALETTE.mocha.colors.blue.hex,
    PALETTE.mocha.colors.green.hex,
    PALETTE.mocha.colors.peach.hex,
    PALETTE.mocha.colors.yellow.hex,
    PALETTE.mocha.colors.red.hex
]

# Map the table colors into a variable called catppucin.
catppuccin_cmap = LinearSegmentedColormap.from_list("catppuccin", catppuccin_colors)

# Create the pivot table
pivot = df.pivot_table(
    index="formatted_work_type",
    columns="remote_allowed",
    values="med_salary",
    aggfunc="median"
)

fig, ax = plt.subplots(figsize=(9, 10))
im = ax.imshow(pivot.values, cmap=catppuccin_cmap) # Use color as mapping

ax.set_xticks(range(len(pivot.columns)))
ax.set_yticks(range(len(pivot.index)))
ax.set_xticklabels(pivot.columns)
ax.set_yticklabels(pivot.index)
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

# Add values
for i in range(len(pivot.index)):
    for j in range(len(pivot.columns)):
        value = pivot.values[i, j]
        ax.text(j, i, f"{value:,.0f}" if not np.isnan(value) else "N/A",
                ha="center", va="center", color="black")

ax.set_title("Median Salary by Employment Type and Remote Status ($)")
fig.tight_layout()
plt.xlabel("Remote Status")
plt.ylabel("Employment Type")
plt.colorbar(im, ax=ax, label="Median Salary (USD)")
plt.tight_layout()
plt.show()
````

# Visualization 4: Distribution of Median Salary

We will now plot 2 different plots based on our median salary. The histograms show the salary distribution across the dataset.

We will use IQR (interquartile range method) to filter out our dataset and remove extreme outliers.


````python
# First Visualization: Distribution of median Salary across entire dataset.
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.hist(df["med_salary"], bins=40, color=PALETTE.mocha.colors.blue.hex, edgecolor="black", alpha=0.8)
plt.xlabel("Median Salary (USD)")
plt.ylabel("Number of Jobs")
plt.title("Overall Salary Distribution")
plt.grid(axis='y', alpha=0.2)


# Filter out outliers using IQR
filtered_df = df[(df['med_salary'] >= Q1 - 1.5 * IQR) & (df['med_salary'] <= Q3 + 1.5 * IQR)]

# Second Visualization: Distribution of median Salary across dataset with outliers removed.
plt.subplot(1,2,2)
plt.hist(filtered_df["med_salary"], bins=40, color=PALETTE.mocha.colors.red.hex, edgecolor="black", alpha=0.8)
plt.xlabel("Median Salary (USD)")
plt.ylabel("Number of Jobs")
plt.title("Overall Salary Distribution (Without Outliers)")
plt.grid(axis='y', alpha=0.2)
plt.tight_layout()
plt.show()
````
