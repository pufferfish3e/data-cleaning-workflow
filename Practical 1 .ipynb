## Question 1:

Write a program that takes an arbitrary number of numerical inputs from the user and determines the largest among them. The program should:

1. Prompt the user to enter the number of values they want to compare.
2. Use a loop to repeatedly ask the user for each numerical input.
3. Track and update the largest number encountered so far.
4. Print the largest number after all inputs have been processed.
   Important Considerations:

-   The program should handle both positive and negative numbers.
-   The program should not store all entered numbers in a list to save memory.
-   The program should utilize a while loop for iteration.

You may only fill up the code in the "Write code here area".


```python
num_count = int(input("How many numbers do you want to enter? "))
largest_number = int() # Initialize with 0
count = 0
while count < num_count:
#Write code here===============================
    num = int(input("Enter a number: "))
    if num > largest_number:
        largest_number = num
    count += 1

###===========================================
print("The largest number is:", largest_number)
```

## Question 2:

Calculate the cGPA (Cumulative Grade Point Average) based on the string scores that has been given to you.

You are provided with the following variables:

a. cGPA that starts as a variable 0.

b. scores as a string that stores 4 different score: 4.0, 3.0, 5.0, 5.0.

Hints:

1. Use str.split()
2. Should type cast be used?

You may only fill up the code in the "Write code here area".


```python
cGPA = 0
scores = "4.0 3.0 5.0 5.0"
#Do not edit the code above

#Write code here=========================
scores_arr = scores.split(" ")
cGPA = sum([int(i[0]) for i in scores_arr]) / len(scores_arr)

#===========================================
print(cGPA)
```
