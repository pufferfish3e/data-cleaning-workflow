# Final Question you will be making a Sleep Tracking Analyzer App


image.png


This is your data called sleep_data with columns for 'Day', 'Deep Sleep (hours)', 'Light Sleep (hours)', 'REM Sleep (hours)', 'Awake Time (hours)', and 'Total Sleep (hours)'. You can use the sample data provided below, but it is important to understand how to replace it with your own sleep tracking data.


```python
sleep_data = pd.DataFrame({
    'Day': ['Mon', 'Tue', 'Wedn', 'Thur', 'Fri', 'Sat', 'Sun'],
    'Deep Sleep (hours)': [2, 2.5, 1.8, 2.2, 2.7, 2, 0],
    'Light Sleep (hours)': [4, 3.5, 3.2, 3.8, 1.3, 1, 0],
    'REM Sleep (hours)': [1.5, 2.8, 1.2, 1.5, 1.7, 2, 1.5],
    'Awake Time (hours)': [0.5, 0.2, 0.8, 0.5, 0.3, 0, 0.5],
    'Total Sleep (hours)': [8, 9, 7, 8, 6, 5, 2]
})
```

Steps to create the dashboard:

1. Bar Chart: Create a bar chart showing the breakdown of sleep stages (Deep Sleep, Light Sleep, REM Sleep, Awake Time) for each day of the week.
2. Pie Chart: Create a pie chart to display the average distribution of sleep stages.
3. Histogram: Create a histogram to visualize the distribution of total sleep hours.
4. Scatter Plot: Create a scatter plot to show the relationship between deep sleep hours and total sleep hours. Add a best-fit line (linear regression) to the scatter plot.
5. Box Plot: Create a horizontal boxplot to visualize the distribution of total sleep hours.
6. Dashboard Layout: Arrange the charts in a 2x3 grid layout such that the bar chart and the pie chart are placed side by side on the first row, with the bar chart occupying the first column and the pie chart occupying the second column, while the histogram, scatter plot, and box plot are placed on the second row, with the histogram occupying the first column, scatter plot occupying the second column, and box plot occupying the third column.

Provide the complete Python code (using only Matplotlib and Pandas) necessary to create this dashboard.


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sleep_data = pd.DataFrame({
    'Day': ['Mon', 'Tue', 'Wedn', 'Thur', 'Fri', 'Sat', 'Sun'],
    'Deep Sleep (hours)': np.array([2, 2.5, 1.8, 2.2, 2.7, 2, 0]),
    'Light Sleep (hours)': np.array([4, 3.5, 3.2, 3.8, 1.3, 1, 0]),
    'REM Sleep (hours)': np.array([1.5, 2.8, 1.2, 1.5, 1.7, 2, 1.5]),
    'Awake Time (hours)': np.array([0.5, 0.2, 0.8, 0.5, 0.3, 0, 0.5]),
    'Total Sleep (hours)':np.array([8, 9, 7, 8, 6, 5, 2])
})
plt.figure(figsize=(16,9))

plt.subplot(2,3,1)
plt.bar(sleep_data["Day"], sleep_data["Deep Sleep (hours)"], label="Deep Sleep")
plt.bar(sleep_data["Day"], sleep_data["Light Sleep (hours)"], label="Light Sleep", bottom=sleep_data["Deep Sleep (hours)"])
plt.bar(sleep_data["Day"],sleep_data["REM Sleep (hours)"], label="REM Sleep", bottom=sleep_data["Light Sleep (hours)"] + sleep_data["Deep Sleep (hours)"])
plt.bar(sleep_data["Day"],sleep_data["Awake Time (hours)"], label="Awake Time", bottom=sleep_data["REM Sleep (hours)"] + sleep_data["Deep Sleep (hours)"] + sleep_data["Light Sleep (hours)"])
plt.legend()
plt.title("Sleep Stages by Day")
plt.xlabel("Day")
plt.ylabel("Hours")

plt.subplot(2,3,2)
labels = ["Deep Sleep (hours)","Light Sleep (hours)","REM Sleep (hours)","Awake time (hours)"]
plt.pie([np.sum(sleep_data["Deep Sleep (hours)"]), np.sum(sleep_data["Light Sleep (hours)"]), np.sum(sleep_data["REM Sleep (hours)"]), np.sum(sleep_data["Awake Time (hours)"])],labels=labels, autopct="%1.1f%%")
plt.title("Average Sleep Stage Distribution")

plt.subplot(2,3,4)
plt.hist(sleep_data["Total Sleep (hours)"])
plt.title("Distribution of Total Sleep Hours")
plt.xlabel("Total Sleep (hours)")
plt.ylabel("Frequency")

plt.subplot(2,3,5)
plt.scatter(sleep_data["Deep Sleep (hours)"],sleep_data["Total Sleep (hours)"])
m, b = np.polyfit(sleep_data["Deep Sleep (hours)"], sleep_data["Total Sleep (hours)"], 1)
plt.plot(sleep_data["Deep Sleep (hours)"],m*sleep_data["Deep Sleep (hours)"]+b,c="r",label="Best-fit line")
plt.title("Deep Sleep vs Total Sleep")
plt.xlabel("Deep Sleep (hours)")
plt.ylabel("Total Sleep (hours)")
plt.legend()

plt.subplot(2,3,6)
plt.boxplot(sleep_data["Total Sleep (hours)"],vert=False,patch_artist=True)
plt.title("Total Sleep Hours Distribution")
plt.xlabel("Total Sleep (hours)")
plt.ylabel("Sleep Stage")
plt.tight_layout()
plt.show()
```
